De la información del data set, podemos ver que las columnas que representan a los jugadores pasados y actuales, así como las personas que tienen acceso al juego y no lo han jugado y aquellas que lo desean jugar tienen que ser números, y están categorizados como stings, por lo que tenemos que transformarlos a strings.

```python
def convert_str_num_to_int(df: pd.DataFrame, col_names: list) -> pd.DataFrame:
    """
    Convierte valores númericos de string a enteros en el dataset.

    Parámetros:
        df (pandas.DataFrame): El dataframe en cuestión.
        col_names (list): Lista de las columnas con los valores a convertir.

    Regresa:
        pandas.DataFrame: Dataframe convertido.
    """
    
    # Copia del dataframe.
    new_df = df.copy()
    
    # Corriendo sobre los nombres de las columnas a cambiar.   
    for col in col_names:
        # Reemplazar 'K' con '000' y '.' con ''
        new_df[col] = new_df[col].str.replace("K", "000").str.replace(".", "")

        # Conversión de valores a enteros
        new_df[col] = new_df[col].astype(int)

    return new_df
```

```python
# Cambiar '[', ']', y "'" En la columna de géneros
data2["Genres"] = data2["Genres"].str.replace("'", "")
data2["Genres"] = data2["Genres"].str.replace("]", "")
data2["Genres"] = data2["Genres"].str.replace("[", "")
data2["Genres"] = data2["Genres"].str.replace(" ", "")

# Se crea una matriz binaria para cada categoría
genres = data2["Genres"].str.get_dummies(",")

# Se suma la recurrencia de cada categoría 
popularity = genres.sum().sort_values(ascending=False)

popularity.head()
```

```python
# Tamaño de la gráfica
fig, ax = plt.subplots(figsize=(10, 5))

# Histograma
sns.barplot(x=popularity[:10].values, y=popularity[:10].index, palette="Blues_r", ax=ax)

# Etiquetas
for i, v in enumerate(popularity[:10].values):
    ax.text(v + 130 / 2, i, str(f"{v} Games"), color="black", ha="center", va="center")

# Título y ejes
ax.set_title("Top 10 géneros de videojuegos más populares", fontsize=15, pad=15)
ax.set_xlabel("")
ax.set_ylabel("")

ax.set_xticklabels([])

ax.tick_params(axis="y", labelsize=11)

sns.despine(right=True, top=True, bottom=True)

plt.show()
```
